```python
import cv2
import os
import numpy as np
from random import shuffle
from tqdm import tqdm
import matplotlib.pyplot as plt
from PIL import Image
from keras.preprocessing.image import load_img, save_img, img_to_array, array_to_img, ImageDataGenerator
```

```python
def Get_Image_Name(img):
  l = len(img)
  label = img[:l-4]
  return label


def augment():
  TRAIN_DIR = "TAKA_DataSet/Training/1000/Box"
  for img in tqdm(os.listdir(TRAIN_DIR)):
    label = Get_Image_Name(img)
    path = os.path.join(TRAIN_DIR, img)
    img = Image.open(path)
    #img = img.rotate(90, expand=True)
    simg = img
    
    img = img_to_array(img)
    #img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    #print(type(img), img.shape)
    samples = np.expand_dims(img, 0)
    #print(samples.shape)
    
    for i in range(5):
        datagen = ImageDataGenerator(width_shift_range=0.2)
        it = datagen.flow(samples, batch_size = 1)
        batch = it.next()
        tmp_image = batch[0].astype('uint8')
        tmp_image = array_to_img(tmp_image)
        
        fileName = 'TAKA_DataSet/Training/1000/' + label + 'WidthShift' + str(i) + '.jpg'
        save_img(fileName, tmp_image)
    
    for i in range(5):
        datagen = ImageDataGenerator(height_shift_range=0.2)
        it = datagen.flow(samples, batch_size = 1)
        batch = it.next()
        tmp_image = batch[0].astype('uint8')
        tmp_image = array_to_img(tmp_image)
        
        fileName = 'TAKA_DataSet/Training/1000/' + label + 'HeightShift' + str(i) + '.jpg'
        save_img(fileName, tmp_image)
        
        
    for i in range(5):
        datagen = ImageDataGenerator(shear_range=12)
        it = datagen.flow(samples, batch_size = 1)
        batch = it.next()
        tmp_image = batch[0].astype('uint8')
        tmp_image = array_to_img(tmp_image)
        
        fileName = 'TAKA_DataSet/Training/1000/' + label + 'Shear' + str(i) + '.jpg'
        save_img(fileName, tmp_image)
        
        
    for i in range(5):
        datagen = ImageDataGenerator(brightness_range=[0.01, 1])
        it = datagen.flow(samples, batch_size = 1)
        batch = it.next()
        tmp_image = batch[0].astype('uint8')
        tmp_image = array_to_img(tmp_image)
        
        fileName = 'TAKA_DataSet/Training/1000/' + label + 'Darken' + str(i) + '.jpg'
        save_img(fileName, tmp_image)
        
    for i in range(5):
        datagen = ImageDataGenerator(brightness_range=[1, 1.9])
        it = datagen.flow(samples, batch_size = 1)
        batch = it.next()
        tmp_image = batch[0].astype('uint8')
        tmp_image = array_to_img(tmp_image)
        
        fileName = 'TAKA_DataSet/Training/1000/' + label + 'Lighten' + str(i) + '.jpg'
        save_img(fileName, tmp_image)
        
    
    for i in range(5):
        datagen = ImageDataGenerator(zoom_range=[.8, 1.5])
        it = datagen.flow(samples, batch_size = 1)
        batch = it.next()
        tmp_image = batch[0].astype('uint8')
        tmp_image = array_to_img(tmp_image)
        
        fileName = 'TAKA_DataSet/Training/1000/' + label + 'zoom' + str(i) + '.jpg'
        save_img(fileName, tmp_image)
    
    
    for i in range(5):
        datagen = ImageDataGenerator(vertical_flip=True)
        it = datagen.flow(samples, batch_size = 1)
        batch = it.next()
        tmp_image = batch[0].astype('uint8')
        tmp_image = array_to_img(tmp_image)
        
        fileName = 'TAKA_DataSet/Training/1000/' + label + 'vert_flip' + str(i) + '.jpg'
        save_img(fileName, tmp_image)
        
        
    for i in range(5):
        datagen = ImageDataGenerator(horizontal_flip=True)
        it = datagen.flow(samples, batch_size = 1)
        batch = it.next()
        tmp_image = batch[0].astype('uint8')
        tmp_image = array_to_img(tmp_image)
        
        fileName = 'TAKA_DataSet/Training/1000/' + label + 'Hor_flip' + str(i) + '.jpg'
        save_img(fileName, tmp_image)
        
    
    
    timg = simg.rotate(180, expand=True)
    fileName = 'TAKA_DataSet/Training/1000/' + label + 'rotate' + str(1) + '.jpg'
    timg.save(fileName)
        
```

```python
augment()
```
